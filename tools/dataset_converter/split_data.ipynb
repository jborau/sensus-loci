```python
import json
import os
import random

# set the path to your data directory
data_path = "/home/javier/datasets/DAIR/single-infrastructure-side-image/"

# set the percentage of data you want to use for validation and testing
val_percent = 0.1
test_percent = 0.1

# create a dictionary to hold the file codes for each split
file_codes = {
    "train": [],
    "val": [],
    "test": []
}

# loop through each file in the data directory
for filename in os.listdir(data_path):
    # extract the file code (assuming it's the first part of the filename before the first underscore)
    file_code = filename.split(".")[0]
    
    # randomly assign the file to the validation, testing, or training split
    rand_num = random.random()
    if rand_num < val_percent:
        file_codes["val"].append(file_code)
    elif rand_num < val_percent + test_percent:
        file_codes["test"].append(file_code)
    else:
        file_codes["train"].append(file_code)

# write the file codes dictionary to a JSON file
with open("data_splits.json", "w") as outfile:
    json.dump(file_codes, outfile)

print(file_codes)
print('Training: ', len(file_codes['train']))
print('Validation: ', len(file_codes['val']))
print('Testing: ', len(file_codes['test']))
```
